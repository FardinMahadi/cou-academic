```python
import numpy as np

def gauss_seidel_method(A, b, initial_guess, tolerance=1e-6, max_iterations=100):
    n = len(b)

    x = initial_guess.copy()
    x_old = np.zeros(n)

    for iteration in range(1, max_iterations + 1):
        x_old = x.copy()
        
        for i in range(n):
            summation_forward = 0.0 
            summation_backward = 0.0 
            
            for j in range(i):
                summation_forward += A[i, j] * x[j]
            
            for j in range(i + 1, n):
                summation_backward += A[i, j] * x_old[j]
            
            x[i] = (b[i] - summation_forward - summation_backward) / A[i, i]

        numerator = np.linalg.norm(x - x_old, np.inf)
        denominator = np.linalg.norm(x, np.inf)

        if numerator / denominator < tolerance:
            print("\nConvergence achieved.")
            print(f"Total iterations: {iteration}")
            return x

    print("\nWarning: Maximum iterations reached without convergence.")
    return x
```

```python
A = np.array([[2.0,1.0],[5.0,7.0]])
b = np.array([11.0,13.0])
guess = np.array([1.0,1.0])

sol = gauss_seidel_method(A,b,guess)
print(sol)
```
