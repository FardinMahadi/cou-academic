```python
def linear_interpolation(x0, y0, x1, y1, x):

    if x0 == x1:
        raise ValueError("x0 and x1 cannot be the same (division by zero).")
    if not (min(x0, x1) <= x <= max(x0, x1)):
        print("Warning: x is outside the interpolation range.")

    y = y0*(x1-x)/(x1-x0) + y1 * (x - x0) / (x1 - x0)
    return y


# Example usage:
x0, y0 = 1, 2
x1, y1 = 4, 8
x = 2.5

y = linear_interpolation(x0, y0, x1, y1, x)
print(f"Interpolated value at x={x} is y={y}")
```
